# Budget Analysis
In this section, you will use the transactions DataFrame to analyze the customer's budget

1. Calculate the total spending per category and print the results (Hint: groupby or count transactions per category)

2. Generate a bar chart with the number of transactions for each category

3. Calulate the expenses per month

4. Plot the total expenses per month

### Calculate the expenses per category

```python
# Compute the total expenses per category
# First, retrieve original transactions and set into new dataframe
df_4 = df_2
df_4.head()
```

```python
# make a new empty dataframe called category_1
category_1 = []
# loop through df_4 for value in a list and append the first value of the list into the empty dataframe, category_1. 
for i in range (0,len(df_4)):
    category_1.append(df_4["category"][i][0])
category_1
```

```python
# set values in category_1 as category column in df_4
df_4["category"] = category_1
df_4.head()
```

```python
# set category as index for df_4 dataframe and drop the duplicated category column
df_4 = df_4.set_index(df_4['category'])
df_4 = df_4.drop(columns=["category"])
df_4.head()
```

```python
# Display the expenses per catefory by grouping the category index and summing the amounts
df_4 = df_4.groupby('category').sum()
df_4
```

```python
# Drop income 
df_expenses = df_4.drop(["Transfer"])
df_expenses
```

```python
# Create a spending categories pie chart.
df_expenses.plot.pie(y="amount")
```
